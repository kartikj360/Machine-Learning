```python
# Step 1: Import the required libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

# Step 2: Load the dataset using pandas read_csv() method
data = pd.read_csv('canada_per_capita_income.csv')

# Step 3: Create the feature matrix (X) and target vector (y)
X = data.iloc[:, :-1].values
y = data.iloc[:, 1].values

# Step 4: Split the dataset into training and testing sets using train_test_split() method
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Step 5: Create a linear regression model using LinearRegression() method
regressor = LinearRegression()

# Step 6: Train the model on the training data using fit() method
regressor.fit(X_train, y_train)

# Step 7: Predict the target values using predict() method
y_pred = regressor.predict(X_test)

# Step 8: Print the predicted value for the year 2020
year_2020 = [[2020]]
predicted_value = regressor.predict(year_2020)
print(predicted_value[0])
```
